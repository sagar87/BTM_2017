# Fun with PCA projections

Load the data.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
X = np.loadtxt('sales_data.txt')
```

Normalize it.

```python
X_norm = (X - np.mean(X, axis =0)) / np.std(X, axis=0)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X_norm[:,0], X_norm[:,1], 'o')
```

The way Harald computes his PCAs...

```python
val, vec = np.linalg.eig(np.cov(X_norm.T))
```

To visualize the vectors along variance is maximized.

```python
h = np.linspace(-2, 2, 100)
v1 = h.reshape(-1, 1) @ vec[:,0].reshape(1, -1)
v2 = h.reshape(-1, 1) @ vec[:,1].reshape(1, -1)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X_norm[:,1], X_norm[:, 0], 'o')
ax.plot(v2[:,0], v2[:, 1], '-')
ax.plot(v1[:,0], v1[:, 1], '-')
```

Now rotate the points.

```python
X_transformed = vec @ X_norm.T
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(X_transformed.T[:,1], X_transformed.T[:, 0], 'o')
ax.vlines(x=0, ymin=ax.get_ylim()[0], ymax=ax.get_ylim()[1], color='orange')
ax.hlines(y=0, xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1], color='green')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
```
